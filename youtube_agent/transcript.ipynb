# Getting ID from a URL

```python
import re

def get_video_id_from_url(url):
    """
    Extracts the YouTube video ID from a given URL.
    """
    # Regex for standard YouTube URLs
    pattern = r"(?:https?:\/\/)?(?:www\.)?(?:youtube\.com|youtu\.be)\/(?:watch\?v=|embed\/|v\/|)([\w-]{11})(?:\S+)?"
    match = re.search(pattern, url)
    if match:
        return match.group(1)
    return None

# Example usage:
youtube_url = "https://youtu.be/y5vE8y_f_OM?si=OlUh5RpFbp8ZX7XL"
video_id = get_video_id_from_url(youtube_url)
if video_id:
    print(f"Video ID: {video_id}")
else:
    print("Could not extract video ID from the URL.")
```

```python
from youtube_transcript_api import YouTubeTranscriptApi

def get_youtube_transcript(video_id):
    """
    Retrieves the transcript for a given YouTube video ID.
    """
    try:
        transcript_list = YouTubeTranscriptApi.get_transcript(video_id)
        transcript_text = ""
        for segment in transcript_list:
            transcript_text += segment['text'] + " "
        return transcript_text.strip()
    except Exception as e:
        return f"Error: {e}"

# Example usage:
transcript = get_youtube_transcript(video_id)
print(transcript)
```

```python
import os

def write_string_to_file(filename, content):
    """
    Writes a given string content to a specified text file.

    Args:
        filename (str): The name of the file to write to (e.g., "my_output.txt").
                        If the file does not exist, it will be created.
                        If the file exists, its content will be overwritten.
        content (str): The string content to be written to the file.
    """
    try:
        # Open the file in write mode ('w').
        # 'w' mode creates the file if it doesn't exist,
        # and overwrites it if it does.
        # 'utf-8' encoding is recommended for general text files
        # to handle various characters.
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(content)
        print(f"Successfully wrote content to '{filename}'")
    except IOError as e:
        print(f"Error writing to file '{filename}': {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# --- Example Usage ---

# 1. Define the string variable you want to write
my_string_variable = "Hello, this is some text that I want to save to a file.\n" \
                     "It can contain multiple lines and special characters like: éàçü."

# 2. Define the filename
output_filename = "output.txt"

# 3. Call the function to write the string to the file
write_string_to_file(output_filename, my_string_variable)

# Optional: Verify the content by reading it back (for demonstration)
print("\n--- Verifying content ---")
try:
    with open(output_filename, 'r', encoding='utf-8') as file:
        read_content = file.read()
    print(f"Content read from '{output_filename}':\n{read_content}")
except FileNotFoundError:
    print(f"Error: File '{output_filename}' not found after writing.")
except IOError as e:
    print(f"Error reading from file '{output_filename}': {e}")

# Optional: Clean up the created file (uncomment if you want to remove it after execution)
# if os.path.exists(output_filename):
#     os.remove(output_filename)
#     print(f"\nCleaned up: Removed '{output_filename}'")
```

```python
from langchain_ollama import ChatOllama

# 1. Initialize your Ollama model
llm = ChatOllama(model="deepseek-r1:8b", temperature=0) # Or "mistral", etc.
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 3. Create a summarization prompt
# Be explicit about what you want the summary to contain (e.g., key points, conciseness)
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an expert summarizer. Summarize the following YouTube video transcript. Focus on the main topics, key information, and any important takeaways."),
        ("user", "Transcript:\n{transcript}\n\nSummary:")
    ]
)

# 4. Chain the prompt, LLM, and an output parser
# The StrOutputParser simply converts the LLM's output message to a string
chain = prompt | llm | StrOutputParser()

# 5. Invoke the chain to get the summary
print("Generating summary...")
summary = chain.invoke({"transcript": transcript})

print("\n--- Generated Summary ---")
print(summary)
```

```python
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "rewrite it, do not remove important info"),
        ("user", "Transcript:\n{transcript}\n\nReport:")
    ]
)

chain = prompt | llm | StrOutputParser()

print("Generating report...")
report = chain.invoke({"transcript": transcript})

print("\n--- Generated Report ---")
print(report)
```
